```python
import codecs
import datetime
from collections import defaultdict
from collections import OrderedDict
import os.path    

#calculates average of wikipedia page views for 30 days before/after movie release
#not optimized ;-)
def calculate_actor_score(text_blob, dt1, dt2):
    actors = str(text_blob).split(";")
    score = 0
    for actor in actors:
        actor_score = 0
        actor_name = actor.replace("*",",")
        filename = 'stars/%s.txt' % (actor_name)
        if os.path.exists(filename):
            #read all days into dictionary
            days_dict = defaultdict()
            with codecs.open(filename, 'r', 'UTF-8') as f:
                for line in f:
                    fields = line.split(",")
                    #date_obj = datetime.strptime(reader[0], '%Y-%M-%D')
                    try:
                        date_obj = datetime.datetime.strptime(fields[0], "%Y-%m-%d")
                        days_dict[date_obj] = int(fields[1])
                    except:
                        #wikipedia dump includes imaginary days
                        pass
                
            #get page views for every day in range
            for day,page_views in days_dict.iteritems():
                if dt1 < day < dt2:
                    #print 'date found' + str(day)
                    score = score + page_views
        else:
            print 'Error file does not exist' + actor_name
    return (score/30) 
            

movies_with_actors['PreOpenScore'] = [calculate_actor_score(row['Actors'], row['Date'] - datetime.timedelta(days=30), row['Date'] ) for index, row in movies_with_actors.iterrows()]
movies_with_actors['PostOpenScore'] = [calculate_actor_score(row['Actors'], row['Date'], row['Date'] + datetime.timedelta(days=30) ) for index, row in movies_with_actors.iterrows()]
```

```python
import matplotlib.patches as mpatches

#makes a historgram of page views around a film
def histogram_page_views(text_blob, dt, movie_title, days_around):
    dt1 = dt - datetime.timedelta(days=days_around)
    dt2 = dt + datetime.timedelta(days=days_around)
    actors = str(text_blob).split(";")
    for actor in actors:
        actor_name = actor.replace("*",",")
        filename = 'stars/%s.txt' % (actor_name)
        if os.path.exists(filename):
            days_dict = defaultdict()
            with codecs.open(filename, 'r', 'UTF-8') as f:
                for line in f:
                    line = line.strip()
                    fields = line.split(",")
                    try:
                        date_obj = datetime.datetime.strptime(fields[0], "%Y-%m-%d").date()
                        if dt1 < date_obj < dt2:
                            #print line
                            days_dict[date_obj] = int(fields[1])
                    except:
                        #wikipedia dump includes imaginary days
                        #print 'error with day ' + fields[0]
                        pass
                    
            od = OrderedDict(sorted(days_dict.items()))
            
            #for k,v in od.items():
            #    print str(k) + ' ' + str(v)
                
            fig, ax = plt.subplots(figsize=(15,5))
            bars  = ax.bar(od.keys(), od.values(), align='center')
            #ax.xticks(range(len(od)), od.keys(), rotation='vertical')
            for i in range(0,days_around-1):
                bars[i].set_facecolor('#C0C0C0')
            bars[days_around-1].set_facecolor('black')
            for i in range(days_around,2*days_around-1):
                bars[i].set_facecolor("#336699")

            ax.xaxis_date()
            
            ax.set_title( str(actors[0]) + ' page views around the release of ' + movie_title )
            fig.autofmt_xdate()
            plt.show()
            
```

```python
def div_boxoffice(op,tot):
    if float(tot) <= 0.01:
        return 0
    else:
        return op/float(tot)
    
def div_opening_buzz(pre,post):
    if float(post) <= 0.01:
        return 0
    else:
        return (post-pre)/float(post)
        
    
movies_with_all['BuzzChange'] = [div_opening_buzz(row['PreOpenScore'],row['PostOpenScore']) for index, row in movies_with_all.iterrows()]
movies_with_all['TotalLessOpening'] = [row['DomesticTotalGross']-row['OpeningWeekend'] for index, row in movies_with_all.iterrows()]
#movies_with_all['TotalLessOpening'] = [div_boxoffice(row['DomesticTotalGross']-row['OpeningWeekend'],row['DomesticTotalGross']) for index, row in movies_with_all.iterrows()]
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std

def train_model(dataframe, explanatory_vars, outcome_var):
    X = dataframe[explanatory_vars]
    Y = dataframe[outcome_var]
    model = sm.OLS(Y, X).fit()
    return model

#Plot the model
plt.scatter(movies_with_all.PreOpenScore, movies_with_all.DomesticTotalGross)
plt.xlabel('Page Views')
plt.ylabel('Gross')
X = movies_with_all[['PreOpenScore', 'Ones']]

model1 = train_model(movies_with_all, ['PreOpenScore', 'Ones'], 'DomesticTotalGross')
plt.plot(movies_with_all.PreOpenScore, model1.predict(X), 'r')

# Plot the 95% confidence intervals
prstd, conf_l, conf_u = wls_prediction_std(model1)
plt.plot(movies_with_all.PreOpenScore, conf_l, 'y')
plt.plot(movies_with_all.PreOpenScore, conf_u, 'y')
```
